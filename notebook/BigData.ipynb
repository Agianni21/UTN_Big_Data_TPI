# Limpieza de datos

Importo dependencias y cargo el csv como Dataframe

```python
import pandas as pd
import numpy as np

df = pd.read_csv("crashes.csv")
df.head()
```

Modifico la columna 'date' para que tenga un formato de fecha mas facil de trabajar

```python

df['date'] = pd.to_datetime(df['date'])
df.head()
```

Reemplazo los valores '?' por celdas vacias (Ayuda al procesamiento en el futuro)

```python
df.replace(to_replace="?", value=np.nan, inplace=True)
df.head()
```

Reemplazo las columnas 'aboard' y 'fatalities' (Strings) por numeros para poder ser procesadas

```python
def extract_value(string):
    import re
    value = re.search("(\d).*\(pass.*", string)
    if value == None:
        value = np.nan
    else:
        value = value.group(1)
    return value
df['aboard'] = df['aboard'].apply(extract_value)
df['fatalities'] = df['fatalities'].apply(extract_value)
df.head()
```

Exporto el dataframe como un nuevos csv limpio

```python
df.to_csv('clean_data.csv')
```
